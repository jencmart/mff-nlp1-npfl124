# 2. Per word Conditional Entropy  $H(Y|X=x), x \in \mathcal{X}$
* this is expected per word conditional entropy
* I also added quantiles q.25 and q.75
* we can see, that per-word conditional entropy is very skewed, because even q.75 is way below the mean
* This confusing table is followed by nice box-plot graphs
    * (because of those outliers, I had to switch off the indication of the outliers ..)

```python
# df.pivot_table(index=["mess_prob"], 
#                 columns=['dataset','experiment'], 
#                 aggfunc=np.mean,
#                 values='word_entropy')
df["MEAN"] = df["word_entropy"]
df["Q25"] = df["word_entropy"]
df["Q75"]  =df["word_entropy"]
df["MAX"] =df["word_entropy"]
p = df.pivot_table(index=["mess_prob"], 
                columns=['dataset','experiment'], 
                aggfunc={'MEAN' : np.mean, 
                         'Q25' : lambda x: np.percentile(x, 25),
                         'Q75' : lambda x: np.percentile(x, 75),
                         'MAX' : np.max},
                #values='word_entropy'
              )

def highlight_max(data, color='yellow'):
    '''
    highlight the maximum in a Series or DataFrame
    '''
    attr = 'background-color: {}'.format(color)
    #remove % and cast to float
    data = data.replace('%','', regex=True).astype(float)
    if data.ndim == 1:  # Series from .apply(axis=0) or axis=1
        is_max = data == data.max()
        return [attr if v else '' for v in is_max]
    else:  # from .apply(axis=None)
        is_max = data == data.max().max()
        return pd.DataFrame(np.where(is_max, attr, ''),
                            index=data.index, columns=data.columns)
p.style.apply(highlight_max)
```
